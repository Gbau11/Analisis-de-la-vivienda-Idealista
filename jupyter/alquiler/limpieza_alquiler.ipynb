```python
sin_ascensor = alquiler['detalles'].str.contains('sin ascensor', case=False, regex=True)

con_ascensor = alquiler['detalles'].str.contains('con ascensor', case=False, regex=True)

chalet = alquiler['detalles'].str.contains('chalet', case=False, regex=True)

alquiler['ascensor'] = ''

alquiler.loc[sin_ascensor, 'ascensor'] = 0

alquiler.loc[con_ascensor, 'ascensor'] = 1

alquiler.loc[chalet, 'ascensor'] = 2
```

```python
condi = ~alquiler['ascensor'].astype(str).isin(['0', '1', '2'])

alquiler.loc[condi, 'ascensor'] = 3

alquiler['ascensor'] = alquiler['ascensor'].astype(int)
```
